```python
import numpy as np
import pandas as pd

np.random.seed(42)
days = 1000

# 1. 模擬正常的股價報酬率 (常態分佈)
returns = np.random.normal(0.05, 2.0, days) # 平均賺 0.05%，波動 2%

# 2. 搞破壞：加入一些髒東西
# 隨機把 50 天的資料變成 NaN (缺漏)
nan_indices = np.random.choice(days, 50, replace=False)
returns[nan_indices] = np.nan

# 隨機加入 10 個超級極端值 (暴跌 20% 或 暴漲 20%)
outlier_indices = np.random.choice(days, 10, replace=False)
returns[outlier_indices] = np.random.uniform(-20, 20, 10)

# 建立 DataFrame
df_analysis = pd.DataFrame({
    'Date': pd.date_range(start='2020-01-01', periods=days),
    'TSMC_Return': returns,
    'Volume': np.random.randint(1000, 50000, days) # 隨機成交量
})

# 故意弄髒成交量：把一些成交量變成負的 (這在現實是不可能的，要抓出來！)
df_analysis.loc[0:5, 'Volume'] = -999 

print("資料已準備好，請用 Data Wrangler 打開 df_analysis！")
```
