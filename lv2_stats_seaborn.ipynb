```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 設定風格
sns.set(style="whitegrid")
np.random.seed(10)

# 1. 產生數據
# 資產 A (穩健): 標準差小，很乖的常態分佈
data_stable = np.random.normal(0, 1, 1000)

# 資產 B (投機): 標準差大，而且混入極端值 (Outliers)
data_risky = np.random.normal(0, 3, 950) # 主體波動就大
outliers = np.random.uniform(-15, 15, 50) # 加入 50 個瘋狂的極端值
data_risky = np.concatenate([data_risky, outliers])

# 整理成 DataFrame 方便畫圖
df = pd.DataFrame({
    'Type': ['Stable Stock'] * 1000 + ['Risky Crypto'] * 1000,
    'Returns': np.concatenate([data_stable, data_risky])
})

print(df)

# 2. 開始畫圖
plt.figure(figsize=(12, 5))

# 左圖：直方圖 (Histogram) - 看分佈形狀
plt.subplot(1, 2, 1)
sns.histplot(data=df, x='Returns', hue='Type', kde=True, bins=30, alpha=0.6)
plt.title('Histogram: Distribution Shape')
plt.xlabel('Daily Returns (%)')

# 右圖：箱型圖 (Boxplot) - 看極端值
plt.subplot(1, 2, 2)
sns.boxplot(data=df, x='Type', y='Returns', palette="Set2")
plt.title('Boxplot: Outlier Detection')
plt.ylabel('Daily Returns (%)')

plt.tight_layout()
plt.show()
# plt.savefig('visual_stat_practice.png')
```
